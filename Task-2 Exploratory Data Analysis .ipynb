```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("Titanic-Dataset.csv")
df
```

### STEP-1 Generate summary statistics (mean, median, std, etc.)

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include='all')
```

```python
print(" Mean Values:\n",df.mean(numeric_only=True))
print("Median Values:\n", df.median(numeric_only=True))
print("Standard Deviation:\n", df.std(numeric_only=True))
print("Missing Values:\n", df.isnull().sum())
```

### STEP-2 Create histograms and boxplots for numeric features.

```python
# Histograms  
numeric_cols = df.select_dtypes(include='number').columns
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=col, bins=30, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
# Boxplots
numeric_cols = df.select_dtypes(include='number').columns
for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col], color='brown')
    plt.title(f"Boxplot of {col}")
    plt.show()
```

### STEP-3 Use pairplot/correlation matrix for feature relationships.


```python
# Optional: clean missing Age or Fare for better plots
df_clean = df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].dropna()

# Pairplot
sns.pairplot(df_clean)
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()
```

```python
# Compute correlation
corr_matrix = df_clean.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

 ### STEP-4 Identify patterns, trends, or anomalies in the data.   

```python
# Survival by passenger class
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
#Survival Rate by Gender:

sns.countplot(data=df, x='Survived', hue='Sex')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Family size analysis
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
plt.figure(figsize=(10, 5))
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title('Survival Rate by Family Size')
plt.show()
```

```python
# Survival by embarkation port
plt.figure(figsize=(8, 5))
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title('Survival Rate by Embarkation Port')
plt.show()
```

### STEP-5 Make basic feature-level inferences from visuals.

```python
print("\nKey Statistics:")
print(f"Overall survival rate: {df['Survived'].mean()*100:.2f}%")
print("\nSurvival rate by class:")
print(df.groupby('Pclass')['Survived'].mean())
print("\nSurvival rate by gender:")
print(df.groupby('Sex')['Survived'].mean())
print("\nAverage fare by class:")
print(df.groupby('Pclass')['Fare'].mean())
print("\nAverage age by survival:")
print(df.groupby('Survived')['Age'].mean())

# Age groups analysis
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 30, 50, 100], 
                      labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])
print("\nSurvival rate by age group:")
print(df.groupby('AgeGroup')['Survived'].mean())
```
